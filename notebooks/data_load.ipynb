```python
import os
import random
import PIL
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
# Define the directories
train_dir = '../data/train'
valid_dir = '../data/valid'
test_dir = '../data/test'
```

```python
def show_sample(dir_nm, n):
    """Displays n-many random images from each class in directory dir_nm"""
    classes = [dir for dir in os.listdir(dir_nm) if not dir.startswith('.')]
    for j in range(n):
        plt.figure(figsize=(7, 4))  # Adjust the figure size if needed
        for i, class_name in enumerate(classes):
            class_dir = os.path.join(dir_nm, class_name) # wildfire, nowildfire
            image_files = os.listdir(class_dir)  # Get all images in the class folder
            random_image = random.choice(image_files)  # Select a random image
            img_path = os.path.join(class_dir, random_image)
            
            # Load the image with its original size
            img = mpimg.imread(img_path)
            
            # Display the image
            plt.subplot(1, len(classes), i + 1)  # Display images in a single row
            plt.imshow(img)
            plt.axis('off')  # Turn off axis for better visualization
            plt.title(f"{class_name} ({img.shape[0]}x{img.shape[1]})")  # Show class name and image size
        plt.show()
```

```python
show_sample(train_dir, 5)
```
